## 1.1 Basic Definitions
- An **Experiment** is any procedure than can be repeated any
number of times (theoretically infinitely), and has a
well-defined set of potential Outcomes.
- An **Outcome**, $\omega$, is a potential eventuality of an Experiment.
- The **Sample Space** or ‘universe’, denoted by $\Omega$, is the set of
all possible outcomes ($\omega$) of the experiment. An outcome is a
member of the sample space, so we can write $\omega \in \Omega$.
- An **Event** $A$ is some subset of $\Omega$, $A \subseteq \Omega$. An elementary
event is one which only contains a single outcome.

## 1.1 Event Example
**Events for Dice**: The sample space for a dice roll
contains six outcomes, i.e. $\Omega_{\text{dice}} = {1, 2, 3, 4, 5, 6}$. We can describe three events $A$, $B$ and $C$ as follows:
- Event $A$ - the rolled number is equal to 6;
- Event $B$ - the rolled number is odd;
- Event $C$ - the rolled number is greater than 3;

Each of these events would then correspond to one or more outcomes in $\Omega_\text{dice}$, namely:

$A = {6} \qquad B = {1, 3, 5} \qquad C = {4, 5, 6}$

## 1.2 Set Theory for Describing Events
- The **union** of $A$ and $B$, denoted $A \cup B$, is the set of those outcomes which belong to $A$ or $B$ (or both). 
    - $A \cup B$ occurs if $A$ occurs or $B$ occurs (or they both occur).

- The **intersection** of $A$ and $B$, denoted $A \cap B$, is the set of those outcomes which belong to both $A$ and $B$. 
    - $A \cap B$ occurs if $A$ occurs and $B$ occurs.

- The **complement** of $A$, denoted $\overline A$, is the set of those outcomes which do not belong to $A$.
  - $\overline A$ occurs if $A$ does not occur.

## 1.2 Note: Subtle Change in Nomenclature
**NOTE**: At this point, we are going to adopt another notation for the **intersection** of two sets. $\color{red}{(A \cap B)}$ is frequently used in the _set theory_ calculations described above. However, in applications of _probability theory_, this is more commonly represented as $\color{blue}{(A, B)}$, which corresponds to a _multivariate_ distribution of two random variables (a subject of importance that will be further explored in subsequent chapters, but for now just note the change in notation).

Long story short: $\color{red}{(A \cap B)}$ will be represented as $\color{blue}{(A, B)}$ from now on when talking about <font color='blue'>Probabilities</font>.

## 1.3 Probability Associated with Events
- Original definitions arose within the context of gambling, or situations where $n$ possible outcomes were equally likely. If some event $A$ were satisfied by $m$ outcomes, then $P(A) = \frac{m}{n}$ (this is the *classical* or *a priori* definition of probability, which we will encounter in subsequent examples).
- However, what if outcomes are not equally likely, or occupy an infinite space? We could then resort to an *empirical* or *a posteriori* definition of probability, where we repeat an experiment *n* times; this method can also suffer from limitations...
- Due to these issues, there was a strong desire to define probability as a *function* rather than a ratio (or limit).

## 1.3 Axioms of Probability
| &thinsp; | &thinsp; | &thinsp; | 
| --- | --- | --- |
| 1st axiom | $P(A) \in \mathbb R$ <br> $\color{blue}{0 \leq P(A) \leq 1}$ | The probability of an event A is a real number between 0 and 1. |
| 2nd axiom | $P(\Omega) = 1$ | The certain event has probability 1.|
| 3rd axiom | If $A \cap B = \emptyset$ <br> then $P(A \cup B) = P(A)+P(B)$ <br> and $\color{blue}{P(\bigcup_{i=1}^{n} A_i) = \sum_{i=1}^{n} P(A_i)}$ | The probability of the union of disjoint events is equal to the sum of their probabilities. |

_Comment on 3rd Axiom:  What if $A \cap B \neq \emptyset$?_ (_See below_)

## 1.3 From Axioms Derive Fundamental Properties
**P1**: $P(\overline A) = 1 - P(A)$

**P2**: $P(\emptyset) = 0$

**P3**: If $A \subset B$ then $P(A) \leq P(B)$

**P4**: $P(A \cup B) = P(A) + P(B) - P(A, B)$
(The inclusion-exclusion formula)

## 1.4 Conditional Probability
- Let $A$ and $B$ be events with $P(B) > 0$. The **conditional probability** $P(A|B)$ (probability of $A$ *given* $B$) is defined as:

$$P(A|B) = \frac{\color{red}{P(A,B)}}{\color{grey}{P(B)}}$$

- This represents the probability that $A$ with occur given that $B$ has occured.

- *Start by thinking about what $P(A,B)$ in the numerator is first...*

## 1.4 Conditional Probability: Example
- What is the probability that a person has seen *Game of Thrones*, given that said person is a *Breaking Bad* fan?
- We want to find $P(G|B)$
$$P(G|B) = \frac{\color{red}{P(G,B)}}{\color{grey}{P(B)}}$$
- Note that this probability is the same as the probability of
selecting a *Game of Thrones* viewer at random from the list of
*Breaking Bad* viewers. Conditional probabilities can be
thought to be reducing the sample space to outcomes
satisfied by the conditioned event only.

```python
Math(r'P(G|B) = {}'.format(BG/(B+BG))) 
```
